```python

# Create a 3x3 array with a specific list of values
arr_list = np.array([[0, 255, 0], [0, 255, 0], [0, 255, 0]])

# Display the image using cv2_imshow
# cv2_imshow expects a single argument: the image data
cv2_imshow(arr_list)
```

```python
resized_image = cv2.resize(arr_list, (300, 300), interpolation=cv2.INTER_NEAREST)

# Display the resized image using cv2_imshow
cv2_imshow(resized_image)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage.color import rgb2gray

# Read the image
img = mpimg.imread('/content/Untitled.jpeg')

# Convert the image to grayscale
gray_img = rgb2gray(img)

# Display the grayscale image
plt.imshow(gray_img, cmap='gray')
plt.show()
```
